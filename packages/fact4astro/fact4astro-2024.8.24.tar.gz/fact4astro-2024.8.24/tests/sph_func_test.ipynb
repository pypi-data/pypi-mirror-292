```python
from astroquery.gaia import Gaia


Gaia.ROW_LIMIT = -1
```

```python
job = Gaia.launch_job_async(
    "select qso.source_name_in_catalogue,qso.catalogue_name,gaia.* from gaiaedr3.agn_cross_id as qso, gaiaedr3.gaia_source as gaia where (qso.catalogue_name='ICRF3 S/X' or qso.catalogue_name='ICRF3 K' or qso.catalogue_name='ICRF3 X/Ka') and gaia.source_id=qso.source_id")
```

```python
gdr2 = job.get_results()
gdr2

# gdr2.show_in_notebook()
```
