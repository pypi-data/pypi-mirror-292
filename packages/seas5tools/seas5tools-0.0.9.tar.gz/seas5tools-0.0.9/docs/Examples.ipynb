```python
import seas5tools
```

```python
seas5 = seas5tools.SEAS5(cdsapi_key='your-api-key')
```

```python
seas5.download('sst','.', year_range=(2024,2024), months=[6], overwrite=True)
```

```python
ds = seas5.proc('.', 'sst', 6, (2024,2024))
```

```python
ds.sel(time='2024-09-01', number=1)['mean'].plot();
```
